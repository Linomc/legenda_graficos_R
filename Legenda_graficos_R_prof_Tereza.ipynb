Estou criando uma pequena base

```r
library(ggplot2)
```

```r
tamanho = c(5,4,5,2,4.5,4,4.5,4,5,4)
length(tamanho)
```

```r
tabela = matrix(tamanho, ncol = 2, byrow = TRUE, dimnames=list(c('a','b','c','d','e'), c('col1','col2')))
```

```r
cores = c("green", "pink")
```

### Gráfico do R apenas

```r
barplot(t(tabela),
        beside = TRUE,
        col = cores,
        ylim = c(0,6),
        axes = FALSE,
        xlab = "abscissa",
        main = "Título",
        )
axis(2, at = 0:5, labels = 0:5)
legend("topright", c('col1','col2') ,col = c("green", "pink"), lwd = 10 )
```
